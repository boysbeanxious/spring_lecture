# 1. 점포 데이터 전처리 

. 기준년, 분기, 상권코드, 서비스업종기준으로 존재하는 점포 데이터  
. 결측치 없음  
. 데이터 unique key 값 : '기준_년_코드','상권_구분_코드', '상권_구분_코드_명', '상권_코드', '상권_코드_명', '서비스_업종_코드'

1. 2020년부터 2021년까지의 데이터를 가지고 있으므로 최종 데이터인 2021년 데이터 선택  
2. 계절성을 없애고 기준년, 상권구분, 상권코드, 서비스 업종 기반으로 분석하기 위해 데이터 전처리(각 점포 상권코드, 서비스업종 코드 별 평균값)

```python
import pandas as pd
import numpy as np
```

```python
df = pd.read_csv('./data/서울시 우리마을가게 상권분석서비스(상권-점포).csv', encoding='cp949')
```

```python
df.head()
```

```python
df.info()
```

### 1. 2020년부터 2021년까지의 데이터를 가지고 있으므로 최종 데이터인 2021년 데이터 선택  

```python
df_1 = df.loc[df['기준_년_코드']==2021]
```

### 2. 계절성을 없애고 기준년, 상권구분, 상권코드, 서비스 업종 기반으로 분석하기 위해 데이터 전처리
. 각 점포 상권코드, 서비스업종 코드 별 평균값

```python
df.columns
```

```python
# 3.3 계절성을 없애고 기준년, 상권구분, 상권코드, 서비스 업종 기반으로 분석하기 위해 데이터 전처리 
jg_groupby_col = ['점포_수', '유사_업종_점포_수', '개업_율', '개업_점포_수',
       '폐업_률', '폐업_점포_수', '프랜차이즈_점포_수']

df_2 = df_1.groupby(['기준_년_코드','상권_구분_코드', '상권_구분_코드_명', '상권_코드', '상권_코드_명',  '서비스_업종_코드', '서비스_업종_코드_명']).mean()[jg_groupby_col].reset_index()
```
