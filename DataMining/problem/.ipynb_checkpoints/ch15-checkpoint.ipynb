```python
from scipy.cluster.hierarchy import dendrogram, linkage, fcluster
from sklearn.cluster import KMeans
from sklearn import preprocessing
from pandas.plotting import parallel_coordinates, scatter_matrix
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
pharm_df = dmba.load_data('Pharmaceuticals.csv')
```

```python
pharm_df.head()
```

```python
pharm_df.info()
```

# a

```python
pharm_df.set_index(['Symbol', 'Name'], inplace=True)
```

```python
df_a = pharm_df.select_dtypes('float64')
```

```python
df_a.head()
```

```python
df_a_scale = df_a.apply(preprocessing.scale, axis = 0)
```

```python
df_a_scale.shape
```

```python
n_num = 2
```

```python
km = KMeans(n_clusters = n_num)
```

```python
model = km.fit(df_a_scale)
```

```python
model.labels_
```

```python
inertia = []

for n_clusters in range(1, 22):
    kmeans = KMeans(n_clusters = n_clusters, random_state = 0).fit(df_a_scale)
    inertia.append(kmeans.inertia_/n_clusters)
    
inertias =  pd.DataFrame({'n_clusters': list(range(1, 22)), 'inertia':inertia})

ax = inertias.plot(x = 'n_clusters', y = 'inertia')
```

```python
inertias
```

# b

```python
plt.rcParams['figure.figsize']  = (8, 8)
fig, axes = plt.subplots(nrows=n_num, ncols=1, constrained_layout=True)

for i in range(n_num):
    axes[i].tick_params(axis='x', labelrotation=45)
    parallel_coordinates(df_a_scale.loc[df_a_scale['label']==i], 'label', ax = axes[i])

```
